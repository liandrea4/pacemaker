# Visualization

```python
import matplotlib.pyplot     as plt
import pylab
import dicom
import json
import os

data_filepath = "/enc_data/eddata/"
img_filepath = "/enc_data/eddata/organized-data/"
```

```python
def get_patient_id(clip_num): 
    for patient_id in os.listdir(img_filepath): 
        for filename in os.listdir(img_filepath + patient_id + "/"): 
            if filename.split('-')[0] == clip_num: 
                return patient_id
    

patients_with_pacemaker = []

with open(data_filepath + "clipnum_with_pacemaker.txt", 'r') as f: 
    for line in f: 
        patient_id = get_patient_id(line.strip())
        if patient_id: 
            patients_with_pacemaker.append(patient_id)
            
with open('data/patientid_with_pacemaker.json', 'w') as f: 
    json.dump(patients_with_pacemaker, f)
```

```python
patient_id = 2525839
filename = "5983771-2197-03-07-0458-ANTERO-POSTERIOR.dcm"
dc = dicom.read_file(img_filepath + str(patient_id) + "/" + filename)

plt.figure(figsize=[10,10])
plt.imshow(dc.pixel_array, cmap=pylab.cm.bone)
plt.show()
```
